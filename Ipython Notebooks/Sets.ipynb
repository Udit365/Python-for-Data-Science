# Sets
---

- Set is a collection of distinct objects, i.e., it doesn't hold any duplicate items.
- Store elements in no particular order.
- Created using curly braces (`{}`)

***Example :***</br>*1. Create a set <b>`age`</b> consisting the age of employees* :


```
age = {56,52,41,63,41}
print(age)
```

*The above output removes the duplicate item(s) from the set and returns a set of unique elements as output.*

*2. Create a set <b>`employee_name`</b> consisting of the names of employees* :

```
employee_name = {"Ram", "Preeti", "Satish","Jhon","Nirmal"}
print(employee_name)
```

## Modifying Sets
---
We can modify sets by using some functions like :
- `add()`
- `discard()`
- `clear()`

### Modify Set Using `add()`
---
- `add()` : Adds element(s) to the existing set at any position.
- ***Syntax :***
> *set_name.add(object)*

***Example :***</br>*Adding the name <b>Ganesh</b> to the existing set <b>`employee_name`</b>* :


```
employee_name.add("Ganesh")
print(employee_name)
```

### Modify Set Using `discard()`
---
- `discard()` : Removes matching objects from an existing set.
- ***Syntax :***
> *set_name.discard(object)*

***Example :***</br>*Dropping the name <b>Jhon</b> from the existing set <b>`employee_name`</b>* :


```
employee_name.discard("Jhon")
print(employee_name)
```

### Modify Set Using `clear()`
---
- `clear()` : Removes all the elements from an existing set.
- ***Syntax :***
> *set_name.clear()*

***Example :***</br>*Clearing the existing set <b>`employee_name`</b>* :

```
employee_name.clear()
print(employee_name)
```

## Set Operations

To perform set operations, let's  create two different sets with different programming languages :


```
Jr_datascientist = {"R", "Python", "SQL"}
Sr_datascientist = {"R","Python", "MDX", "Javascript", "SQL", "DAX"}

print(Jr_datascientist)
print(Sr_datascientist)
```

### Set Union
---
- `union()` : Returns all the elements belonging to both `Set-A` and `Set-B`.
- This is `[A u B]`
- ***Syntax :***
> *set_A.union(set_B)*

***Example :***

```
Union = Jr_datascientist.union(Sr_datascientist)
print(Union)
```

### Set Intersection
---
- `intersection()` : Returns all the elements common to `Set-A` and `Set-B`.
- This is `[A n B]`
- ***Syntax :***
> *set_A.intersection(set_B)*

***Example :***

```
Intersection = Jr_datascientist.intersection(Sr_datascientist)
print(Intersection)
```

### Set Difference
---
- `difference()` : Returns elements belonging to `Set-A` but not `Set-B`.
- This is `[A-B]`
- ***Syntax :***
> *set_A.difference(set_B)*

***Example :***

```
Difference = Jr_datascientist.difference(Sr_datascientist)
print(Difference)
```

### Symmetric Difference
---
- `symmetric_difference()` : Returns elements not common to both sets
- This is `1-[A n B]`
- ***Syntax :***
> *set_A.symmetric_difference(set_B)*

***Example :***

```
Symmetric_Diff = Jr_datascientist.symmetric_difference(Sr_datascientist)
print(Symmetric_Diff)
```
